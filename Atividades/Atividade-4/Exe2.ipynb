```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# dados de treinamento 
temp = np.array([100,200,300,400,500,600,700]).reshape(-1,1) #Temperatura 
qtd = np.array([30,35,40,45,60,100,150]) # Quantidade de calor fornecido

#Criar e treinar o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(temp,qtd)

#dados de teste 

#qtd_teste = np.array([30,35,40,45,60,100,150]).reshape(-1,1) #vazão de água em m3/min
temp_teste = np.array([100,200,300,400,500,600,700]).reshape(-1,1) #vazão de água em m3/min

#fazer previsões com base na vazão de água
qtd_calor = modelo.predict(temp_teste)
```

```python
import matplotlib.pyplot as  plt
```

```python
#imprimir as previsões
for i in range(len(temp_teste)):
    print(f"Temperatura: {temp_teste [i][0]} C° - Quantidade de calor forncecida: {qtd_calor[i]} " )
plt.scatter(temp,qtd, color='blue', label ='Dados de treinamento')
plt.plot(temp_teste,qtd_calor, color = 'red', label = 'Dados de teste')
plt.xlabel('Temperatura')
plt.ylabel('Quantidade de calor fornecido')
plt.title('Previsão de temperatura a partir da quantidade de calor fornecido')
plt.legend()#exibe a legenda do gráfico
plt.grid(True)#exibe o grid do gráfico
plt.show()#exibe o gráfico
```
