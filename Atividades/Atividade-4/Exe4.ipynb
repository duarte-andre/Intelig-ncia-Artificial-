```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
#inserir dados 
Renda = np.array([1200,5000,8000,10000,30000]).reshape(-1, 1)
gastos = np.array([750,1500,4000,6000,18000])

# Criar e treinar o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(Renda, gastos)

# Dados de teste
Renda_teste = np.array([1200,5000,8000,10000,30000]).reshape(-1, 1)  # tamanho da área

# Fazer previsões com base nas áreas
gastos_renda = modelo.predict(Renda_teste)
```

```python
import matplotlib.pyplot as  plt
```

```python
#imprimir as previsões
for i in range(len(Renda_teste)):
    print(f"Renda: {Renda_teste [i][0]} $ - Gasto renda: {gastos_renda[i]} $" )
plt.scatter(Renda,gastos, color='blue', label ='Dados de treinamento')
plt.plot(Renda_teste,gastos_renda, color = 'red', label = 'Dados de teste')
plt.xlabel('Renda')
plt.ylabel('Gastos')
plt.title('Relação renda com gastos')
plt.legend()#exibe a legenda do gráfico
plt.grid(True)#exibe o grid do gráfico
plt.show()#exibe o gráfico
```
