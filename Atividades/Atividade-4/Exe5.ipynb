```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
from sklearn import tree
import seaborn as sns
from sklearn import datasets
from sklearn import tree
```

```python
# Dados de treinamento
consumo = np.array([50, 140, 30, 75, 123]).reshape(-1, 1)  # Consumo em KW
demanda = np.array([300, 900, 200, 550, 800])  # Demanda de energia correspondente ao consumo

# Criar e treinar o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(consumo, demanda)

# Dados de teste (consumo dos meses anteriores)
consumo_teste = np.array([60, 120, 40, 80, 130]).reshape(-1, 1)

# Fazer previsões de demanda com base nos consumos anteriores
demanda_prevista = modelo.predict(consumo_teste)
```

```python
import matplotlib.pyplot as  plt
```

```python
#imprimir as previsões
for i in range(len(consumo_teste)):
    print(f"Consumo: {consumo_teste [i][0]} KW - Demanda: {demanda_prevista[i]} KW/H Mês" )
plt.scatter(consumo,demanda, color='blue', label ='Dados de treinamento')
plt.plot(consumo_teste,demanda_prevista, color = 'red', label = 'Dados de teste')
plt.xlabel('Consumo em KW/H')
plt.ylabel('Demanda em KW/H')
plt.title('Relação de consumo e demanda')
plt.legend()#exibe a legenda do gráfico
plt.grid(True)#exibe o grid do gráfico
plt.show()#exibe o gráfico
```
