```python
import numpy as np 
from sklearn.linear_model import LinearRegression
```

```python
# Dados de treinamento
A = np.array([50, 100, 120, 300]).reshape(-1, 1)  # área
V = np.array([180.000, 300.000, 375.000, 600.000])  # volume

# Criar e treinar o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(A, V)

# Dados de teste
Area_teste = np.array([50, 100, 120, 300]).reshape(-1, 1)  # tamanho da área

# Fazer previsões com base nas áreas
valor_area = modelo.predict(Area_teste)
```

```python
import matplotlib.pyplot as  plt
```

```python
#imprimir as previsões
for i in range(len(Area_teste)):
    print(f"Área: {Area_teste [i][0]} m² - Valor da área: {valor_area[i]} $" )
plt.scatter(A,V, color='blue', label ='Dados de treinamento')
plt.plot(Area_teste,valor_area, color = 'red', label = 'Dados de teste')
plt.xlabel('Área (m²)')
plt.ylabel('Valor da área')
plt.title('Relação de área e valor')
plt.legend()#exibe a legenda do gráfico
plt.grid(True)#exibe o grid do gráfico
plt.show()#exibe o gráfico
```
